```python
import pandas as pd 
import numpy  as np 
import re
```

```python
data = pd.read_csv(r"BL-Flickr-Images-Book.csv")
data.head
```

```python
data.columns
```

```python
to_drop = ['Former owner', 'Edition Statement', 'Engraver', 'Issuance type', 'Corporate Author', 'Corporate Contributors', 'Shelfmarks']
data1 = data.copy()
```

```python
data1 = data1.drop(columns=to_drop)
data1
```

```python
data2 = data1.set_index('Identifier')
data2
```

```python
newColumn = data2['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
newColumn
```

```python
data2['Date of Publication'] = pd.to_numeric(newColumn)
data2
```

```python
place = data2['Place of Publication']
london = place.str.contains('London')
london
```

```python
oxford = place.str.contains("Oxford")
```

```python
data2['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               place.str.replace('-', ' ')))
```

```python
unwanted_characters = ['[', ',', '-',']']
def clean_dates(item):
    dop = item.get('Date of Publication')  # Safe way to access column data
    if pd.isna(dop):  # Handle actual NaN
        return np.NaN
    
    dop = str(dop)
    if dop.startswith('['):  # Check for unwanted start
        dop = dop[1:]
    
    for character in unwanted_characters:
        dop = dop.split(character)[0]  # Split on the character and take the first part
    
    return dop.strip()  # Clean trailing spaces

# Apply function to the DataFrame
data2['Date of Publication'] = data2.apply(clean_dates, axis=1)
data2
```

```python


def clean_author_names(author):
    # Handle missing values
    if pd.isna(author):
        return 'NaN'
    
    # Convert to string
    author = str(author).strip()
    
    # Handle 'nan' as a string
    if author.lower() == 'nan':
        return 'NaN'
    
    # Split by comma
    parts = author.split(',')
    
    # If no comma, assume it's a single name
    if len(parts) == 1:
        name = parts[0]
        # Remove special characters but preserve spaces and hyphens
        name = re.sub(r'[^a-zA-Z\s-]', '', name)
        return name.strip()
    
    # Extract last name and first name
    last_name, first_name = parts[0].strip(), parts[1].strip()
    
    # Clean first name
    # Remove anything after a hyphen (optional, depending on your needs)
    if '-' in first_name:
        first_name = first_name[:first_name.find('-')]
    
    # Remove trailing periods or special characters
    first_name = re.sub(r'[\.\|]+$', '', first_name)
    
    # Capitalize last name
    last_name = last_name.capitalize()
    
    # Combine and return
    return f'{first_name.strip()} {last_name.strip()}'

# Apply the function to the DataFrame
data2['Author'] = data2['Author'].apply(clean_author_names)
data2
```

```python
def clean_title(title):
    # Handle missing values
    if pd.isna(title):
        return 'NaN'
    
    # Convert to string
    title = str(title).strip()
    
    # Handle 'nan' as a string
    if title.lower() == 'nan':
        return 'NaN'
    
    # Remove content inside brackets
    if title.startswith('['):
        title = title[1:title.find(']')]
    
    # Remove everything after 'by' or 'By' (case-insensitive)
    by_index = title.lower().find('by')
    if by_index != -1:
        title = title[:by_index].strip()
    
    # Remove additional bracketed content
    bracket_index = title.find('[')
    if bracket_index != -1:
        title = title[:bracket_index].strip()
    
    # Remove trailing special characters (optional)
    title = title.strip('.,!?;:- ')
    
    # Capitalize words (using title() for better formatting)
    title = title.title()
    
    return title

# Apply the function to the DataFrame
data2['Title'] = data2['Title'].apply(clean_title)
data2.head()
```

```python
# Define a dictionary of replacements
place_mapping = {
    r'(?i)london': 'London',  # (?i) makes the regex case-insensitive
    r'(?i)oxford': 'Oxford',
    r'(?i)newcastle upon tyne': 'Newcastle-upon-Tyne'
}

# Use replace() with regex=True
data2['Place of Publication'] = data2['Place of Publication'].replace(place_mapping, regex=True)
data2
```

```python
data2.to_csv("Final output.csv" ,index=False)
```
